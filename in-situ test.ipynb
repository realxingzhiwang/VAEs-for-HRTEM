```python
def radial_profile(data, center):
    y, x = np.indices((data.shape))
    r = np.sqrt((x - center[0])**2 + (y - center[1])**2)
    r = r.astype(np.int)

    tbin = np.bincount(r.ravel(), data.ravel())
    nr = np.bincount(r.ravel())
    radialprofile = tbin / nr
    
    all_r = np.arange(r.min(), r.max()+1, 1)
    non_nan = radialprofile==radialprofile
    
    return radialprofile[non_nan], all_r[non_nan]
```

```python
def angular_profile(data, center, r_range):
    y, x = np.indices((data.shape))
    r = np.sqrt((x - center[0])**2 + (y - center[1])**2)
    theta = np.arctan((y - center[1])/(x - center[0]+1e-10))/np.pi*180+90
    theta = theta.astype(np.int)
    
    theta_flat = theta.ravel()
    data_flat = data.ravel()
    r_flat = r.ravel()
    
    roi = np.logical_and(r_flat>r_range[0], r_flat<r_range[1])

    tbin = np.bincount(theta_flat[roi], data_flat[roi])
    ntheta = np.bincount(theta_flat[roi])
    angularprofile = tbin / ntheta
    
    all_theta = np.arange(theta_flat[roi].min(), theta_flat[roi].max()+1, 1)
    non_nan = angularprofile==angularprofile
    
    return angularprofile[non_nan], all_theta[non_nan]
```

```python
radialprofile, r = radial_profile(magnitude_spectrum,(2048,2048))
plt.plot(r[170:230], radialprofile[170:230])
```

```python
angularprofile, theta = angular_profile(magnitude_spectrum, (2048,2048), (203,210))
exponantial = np.exp(angularprofile-np.min(angularprofile))
plt.plot(theta, angularprofile)
plt.xlabel('Theta (degree)')
plt.ylabel('Intensity')
```

```python
peaks, properties = find_peaks(exponantial, height=exponantial.max()/100000)
if exponantial[0]>=exponantial.max()/100:
    peaks = np.insert(peaks, 0, theta[0])
elif exponantial[0]>=exponantial.max()/100:
    peaks = np.append(peaks, theta[-1])
fwhm = peak_widths(exponantial, peaks, rel_height=0.5)
fwhm = np.round(fwhm[0])
fwhm[fwhm<1] = 1
```

```python
peaks
```

```python
def gaussian2d(size, mu, sigma):
    x, y = np.meshgrid(np.arange(0, size[0], 1), np.arange(0, size[1], 1))
    return np.exp(-((x-mu[0])**2/(2*sigma[0]**2)+(y-mu[1])**2/(2*sigma[1]**2)))
```

```python
idx = list()
for i in range(sum(peaks<=90)):
    peak_range = np.arange(peaks[i]-fwhm[i], peaks[i]+fwhm[i]+1, 1)+90
    if np.isin(peak_range, peaks[peaks>=90]).any():
        idx.append(i)
```

```python
idx
```

```python
center = (2048,2048)
centroids = list()
for n in idx:
    selected_peaks = np.zeros(dft_shift.shape)
    for i in range(1024+512,3072-512):
        for j in range(1024+512,3072-512):
            r = np.sqrt((i-center[0])**2+(j-center[1])**2)
            theta = np.arctan((j - center[1])/(i - center[0]+1e-10))/np.pi*180+90
            if 170<r<230 and (peaks[n]-fwhm[n]<theta<peaks[n]+fwhm[n] or peaks[n]-fwhm[n]+90<theta<peaks[n]+fwhm[n]+90):
                selected_peaks[i,j] = dft_shift[i,j]
                    
    f_ishift = np.fft.ifftshift(selected_peaks)
    im_back = np.abs(cv2.idft(f_ishift))
    im_back = cv2.magnitude(im_back[:,:,0],im_back[:,:,1])
    plt.figure()
    plt.imshow(cv2.magnitude(selected_peaks[1024+512:3072-512,1024+512:3072-512,0],selected_peaks[1024+512:3072-512,1024+512:3072-512,1]))
    centroids.append(np.unravel_index(np.argmax(im_back), im_back.shape))
```

```python
center = np.array([2048,2048])
centroids = list()
R = 205
for n in idx:
    center1 = np.array([R*np.sin((peaks[n]-90)/180*np.pi), R*np.cos((peaks[n]-90)/180*np.pi)]).round()
    center2 = np.array([R*np.sin((peaks[n])/180*np.pi), R*np.cos((peaks[n])/180*np.pi)]).round()
    center3 = np.array([-R*np.sin((peaks[n]-90)/180*np.pi), -R*np.cos((peaks[n]-90)/180*np.pi)]).round()
    center4 = np.array([-R*np.sin((peaks[n])/180*np.pi), -R*np.cos((peaks[n])/180*np.pi)]).round()
    kernel = gaussian2d(center*2, center+center1, (5,5))+gaussian2d(center*2, center+center2, (5,5))+gaussian2d(center*2, center+center3, (5,5))+gaussian2d(center*2, center+center4, (5,5))
    selected_peaks = dft_shift*np.stack((kernel, kernel), axis=-1)
                    
    f_ishift = np.fft.ifftshift(selected_peaks)
    im_back = np.abs(cv2.idft(f_ishift))
    im_back = cv2.magnitude(im_back[:,:,0],im_back[:,:,1])
    plt.figure()
    plt.imshow(cv2.magnitude(selected_peaks[1024+512:3072-512,1024+512:3072-512,0],selected_peaks[1024+512:3072-512,1024+512:3072-512,1]))
    #plt.imshow(im_back)
    centroids.append(np.unravel_index(np.argmax(im_back), im_back.shape))
```
